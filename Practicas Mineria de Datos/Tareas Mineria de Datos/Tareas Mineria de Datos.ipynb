### 4.El objetivo es una clasificación con los datos, asi que hay que acomodar los datos de manera que se nos facilite esta tarea.

```python
#Creamos una clase, despues creamos una nueva clasificacion, el cual va a estar dada por las peliculas que tengan imdb_score igual 
#a 7.5, genera un nuevo CSV el cual solo guardara todas las peliculas con imdb_score = 7.5
clase_1 = df[df['imdb_score']==7.5]

clase_1.to_csv('clase1.csv', index=False)
```
